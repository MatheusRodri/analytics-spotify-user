## 🥉 Bronze Layer
Reading JSON files and writing to Delta table.

```python
# Execute the setup script to prepare the environment and load necessary libraries
%%capture
%run ./00-Setup
```

```python
# Read the JSON files containing the streaming history data into a Spark DataFrame
df_input = spark.read\
    .format("json")\
    .option("multiline", "true")\
    .option("inferSchema", "true")\
    .load(PATH_ORIGIN + "Streaming_History_*.json")
```

```python
# Add ingestion timestamp and source file path metadata to the DataFrame
df_bronze = df_input.select(
    "*", 
    current_timestamp().alias("dt_ingestion"), 
    col("_metadata.file_path").alias("source_file") 
    )
```

```python
# Save the dataframe in Bronze Layer
df_bronze.write.format("delta").mode("overwrite").option("mergeSchema", "true").saveAsTable(NAME_TABLE_BRONZE)
print(f"✅ Loaded table: {NAME_TABLE_BRONZE}")
```
