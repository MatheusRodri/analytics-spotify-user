## Import Libs

```python
import pandas as pd
import glob
from dotenv import load_dotenv
import os
from pyspark.sql.functions import col
```

## Load Env

```python
load_dotenv()
```

## Read Silver Table

```python
df_audio_01 = spark.table(os.getenv("NAME_TABLE_AUDIO_SILVER"))
```

```python
display(df_audio_01.limit(5))
```

```python
df_audio_02 = df_audio_01.select(
    col("nr_ms_played"),
    col("nm_track")
)
display(df_audio_02.limit(20).sort("nm_track"))
```

```python
df_audio_02.createTempView("df_audio_02")
```

```python
df_audio_03 = spark.sql('''
                        select 
                        nm_track,
                        max(nr_ms_played) as max_ms_played
                        from df_audio_02
                        group by nm_track
                         ''')
```

## Save dataframe

```python
df_audio_03.write.mode("overwrite").saveAsTable(os.getenv("NAME_TABLE_AUDIO_SILVER_DISTINC"))
```
