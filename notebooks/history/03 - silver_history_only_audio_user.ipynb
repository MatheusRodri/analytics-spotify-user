## Import Libs

```python
import pandas as pd
import glob
from dotenv import load_dotenv
import os
from pyspark.sql.functions import col
```

## Load Env

```python
load_dotenv()
```

## Read Silver Table

```python
df_audio_01 = spark.table(f"{os.getenv("CATALOG")}.{os.getenv("SILVER_SCHEMA")}.{os.getenv("SILVER_NAME_TABLE_AUDIO")}")
```

```python
df_audio_02 = df_audio_01.select(
    col("nr_ms_played"),
    col("nm_track")
)
```

```python
df_audio_02.createOrReplaceTempView("df_audio_02")
```

```python
df_audio_03 = spark.sql('''
                        select distinct
                        nm_track,
                        max(nr_ms_played) as max_ms_played
                        from df_audio_02
                        group by nm_track
                         ''')
```

## Save dataframe

```python
df_audio_03.write.mode("overwrite").saveAsTable(f"{os.getenv("CATALOG")}.{os.getenv("SILVER_SCHEMA")}.{os.getenv("SILVER_NAME_TABLE_AUDIO_DISTINCT")}" )
```
