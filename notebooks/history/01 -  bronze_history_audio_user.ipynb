## Import Libs

```python
import pandas as pd
import glob
from dotenv import load_dotenv
import os
```

## Load Env

```python
load_dotenv()
```

## Read File

```python
# Read all Audio history files and concatenate
files = glob.glob("../data/extended/Streaming_History_Audio_*.json")

df_audio_concat = pd.concat([pd.read_json(file) for file in files], ignore_index=True)
```

## Convert pandas in spark dataframe

```python
df_audio = spark.createDataFrame(df_audio_concat)
```

## Save dataframe

```python
df_audio.write.mode("overwrite").saveAsTable(f"{os.getenv("CATALOG")}.{os.getenv("BRONZE_SCHEMA")}.{os.getenv("BRONZE_NAME_TABLE_AUDIO")}")
```
