```python
from databricks.sdk import WorkspaceClient
```

```python
# Capture host
host = spark.conf.get("spark.databricks.workspaceUrl")
#Capture HTTP Path
w = WorkspaceClient()
```

```python
print(f"Your host is: {host} <- copy this value")

# List warehouses to find the HTTP Path for Power BI connection
for dw in w.warehouses.list():
    print(f"HTTP Path: {dw.odbc_params.path}  <- copy this value")
```
